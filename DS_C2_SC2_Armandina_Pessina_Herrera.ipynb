Se importan las librerías de Pandas y NumPY

```python
import pandas as pd
import numpy as np
```

Se abre el archivo para acceder a la información que contiene y se informa si se logró tener acceso

# a. Extraer la información del archivo Dataframe=df

```python
df = pd.read_csv('country_vaccinations.csv')
print("Archivo cargado exitosamente.")
```

# b. Mostrar estructura y tipos de datos (Asegurando datetime64)

# Convertimos la columna 'date' a datetime64

```python
df['date'] = pd.to_datetime(df['date'])
print("\n--- Estructura y Tipos de Datos ---")
print(df.info())
```

# c. Cantidad de vacunas aplicadas por compañía (agrupado por columna 'vaccines')

# Nota: Usamos 'daily_vaccinations' ya que 'total_vaccinations' es acumulativo

```python
vacunas_por_compania = df.groupby('vaccines')['daily_vaccinations'].sum()
print("\n--- Vacunas por combinaciones de compañías ---")
print(vacunas_por_compania.head())
```

# d. Cantidad de vacunas aplicadas en todo el mundo

```python
total_mundo = df['daily_vaccinations'].sum()
print(f"\nTotal de vacunas aplicadas en el mundo: {total_mundo:,.2f}")
```

# e. Promedio de vacunas aplicadas por país

```python
promedio_por_pais = df.groupby('country')['daily_vaccinations'].mean()
print("\n--- Promedio de vacunas por país (Top 5) ---")
print(promedio_por_pais.head())
```

# f. Cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo

```python
fecha_especifica = '2021-01-29'
vacunas_fecha = df[df['date'] == fecha_especifica]['daily_vaccinations'].sum()
print(f"\nVacunas aplicadas el {fecha_especifica}: {vacunas_fecha:,.2f}")
```

# g. Crear dataframe 'conDiferencias' (daily_vaccinations - daily_vaccinations_raw)

```python
conDiferencias = df.copy()
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
```

# h. Periodo de tiempo entre el registro más reciente y el más antiguo

```python
diferencia_tiempo = df['date'].max() - df['date'].min()
print(f"\nPeriodo de tiempo total en el registro: {diferencia_tiempo.days} días")
```

# i. Crear dataframe 'conCantidad' (canVac: cantidad de vacunas usadas por registro)

```python
conCantidad = df.copy()
# Contamos cuántas vacunas hay separando por coma
conCantidad['canVac'] = conCantidad['vaccines'].str.split(',').str.len()
```

# j. Generar dataframe 'antes20' (registros antes del 20 de diciembre de 2020)

```python
antes20 = df[df['date'] < '2020-12-20']
```

# k. Obtener dataframe 'pfizer' (donde se haya utilizado la vacuna Pfizer)

```python
pfizer = df[df['vaccines'].str.contains('Pfizer', case=False)]
```

# l. Almacenar dataframes en un archivo Excel con múltiples hojas

# ----------------------------------------------------------------

```python
with pd.ExcelWriter('resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)

print("\n¡Archivo 'resultadosReto.xlsx' generado con éxito!")
```
